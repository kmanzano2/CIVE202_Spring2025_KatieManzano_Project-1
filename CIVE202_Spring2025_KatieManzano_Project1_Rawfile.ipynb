```python
import pandas as pd
```

```python
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')
air_data = pd.DataFrame(air_data)
```

```python
#Task 1
air_data_grouped = air_data.groupby('sensor.name').agg(
    VOC_mean =('voc', 'mean'),
    VOC_median=('voc', 'median'),
    pm25_mean=('pm2.5_atm', 'mean'),
    pm25_median=('pm2.5_atm', 'median'),
    pm10_mean=('pm10.0_atm', 'mean'),
    pm10_median=('pm10.0_atm', 'median')

)

#PM2.5
top5_stat_pm25 = (
    air_data_grouped.sort_values(by=['pm25_mean','pm25_median'], ascending = False)
)
top5_stat_pm25.head(5)
#PM10.0
top5_stat_pm10 = (
    air_data_grouped.sort_values(by=['pm10_mean','pm10_median'], ascending = False)
)
#VOC
top5_stat_voc = (
    air_data_grouped.sort_values(by=['VOC_mean','VOC_median'], ascending = False)
)
top5_stat_voc.head(5)
```

```python
#Task 2
unhealthy_voc = air_data[air_data['voc']>=25]
unhealthy_events = unhealthy_voc[
    ['date', 'sensor.name', 'voc']
].sort_values('date')
unhealthy_events.head(10)
```

```python
#Task 3
def assign_temperature(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32<=temperature<=50: 
        return 'Cool'
    elif 51<=temperature<=70:
        return 'Warm'
    else:
        return 'Hot'
air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)
def assign_hum(humidity):
    if humidity <50:
        return 'Low'
    elif 50<=humidity<=80:
        return 'High'
    else:
        return 'Very High'
        
air_data['humidity_level'] = air_data['humidity'].apply(assign_hum)
hum_temp_air_data = air_data.groupby(
    ['humidity_level', 'temperature_level']
).agg(
     VOC_mean =('voc', 'mean'),
    pm25_mean=('pm2.5_atm', 'mean'),
    pm10_mean=('pm10.0_atm', 'mean')
)
air_data.head(10)

```
